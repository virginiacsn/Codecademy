```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('all_data.csv')
print(list(data.columns))

data.rename(columns={'Life expectancy at birth (years)': 'Expectancy'}, inplace=True)
data['Country'] = data['Country'].replace('United States of America', 'USA')
print(data.head())

print(data.dtypes)
```

```python
print(data.isna().sum())
```

```python
print(data.groupby('Country')['Expectancy'].describe())
```

```python
# Plot life expectancy and GDP in time
fig, ax = plt.subplots(1, 2, figsize=(8, 3))
sns.lineplot(data=data, x='Year', y='Expectancy', 
            hue='Country', ax=ax[0], legend=False)
sns.lineplot(data=data, x='Year', y='GDP', 
            hue='Country', ax=ax[1], legend='brief')
plt.legend(bbox_to_anchor=(1, 0.8), frameon=False)
plt.tight_layout(pad=0.5)
for axi in ax:
    axi.spines['top'].set_visible(False)
    axi.spines['right'].set_visible(False)
plt.show()
```

```python
sns.catplot(data=data, x='Country', y='Expectancy', hue='Country',
            kind='point', linestyle='none', height=3.5, aspect=1.5)
sns.stripplot(data=data, x='Country', y='Expectancy', hue='Country', 
              size=3, alpha=0.5)
plt.xticks(rotation=45)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
sns.scatterplot(data=data, x='Expectancy', y='GDP', hue='Country', ax=ax)
plt.legend(bbox_to_anchor=(1, 0.8), frameon=False)
plt.tight_layout(pad=0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```
